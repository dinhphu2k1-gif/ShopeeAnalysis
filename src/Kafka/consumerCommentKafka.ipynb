```python
from kafka import KafkaProducer
import requests
from json import dumps
import time
from kafka import KafkaConsumer
from json import loads
import pandas as pd
```

```python
consumer = KafkaConsumer(
    'Comment',
     bootstrap_servers=['127.0.0.1:9092'],
     auto_offset_reset='earliest',
     enable_auto_commit=True,
     group_id='my-group',
     value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
from pyspark import SparkContext, SparkConf

conf = SparkConf().setAppName("Upload Comment to HDFS").setMaster("spark://25.15.27.228:7077")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Upload Comment to HDFS').getOrCreate()
```

```python
df = []
iter = 0
for message in consumer:
    iter +=1
    message = message.value
    df.append(message)
    if iter %10000 ==0:
        iter = 0
        print(len(df))
    if len(df) > 40000:
        print('Upload Comment to HDFS')
        try:
            pandas_df = pd.DataFrame(df)
            df = []
            data = spark.createDataFrame(pandas_df.astype(str))
            data.coalesce(1).write.mode('append').parquet('hdfs://cris:9000/shopee/Comment/')
        except:
            print('Error when upload data to HDFS')
```
